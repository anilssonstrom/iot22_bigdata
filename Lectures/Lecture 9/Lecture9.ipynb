```
sc = spark.sparkContext.getOrCreate()
sc
```

```
myrange = range(16)
```

```
rdd = sc.parallelize(myrange)
rdd.cache()
```

```
print(rdd)
```

```
print(rdd.map(lambda x: x*2))
```

```
def square_fun(x):
    return x ** 2

square_lambdafun = lambda x: x ** 2
```

```
print(square_fun(4))
print(square_lambdafun(4))
```

```
def add_fun(a, b):
    return a + b

add_lambdafun = lambda a, b: a + b

print(add_fun(4, 3))
print(add_lambdafun(4, 3))
```

```
rdd_filtered = rdd.filter(lambda x: x % 2 == 0).collect()
print(rdd_filtered)
```

```
rdd_kv = sc.parallelize([('a', 1), ('b', 0), ('b', 2), ('a', 5)])
```

```
rdd_kv.reduceByKey(lambda x, y: x + y).collect()
```

```
countletter = sc.parallelize([('a', 1), ('b', 6), ('c', 2), ('d', 5)])
```

```
defletter = sc.parallelize([('a', 'vowel'), ('b', 'consonant'), ('c', 'consonant'), ('d', 'consonant')])
```

```
countletter.join(defletter).map(lambda x: (x[1][1], x[1][0])).reduceByKey(lambda x, y: x + y).collect()
```

```
countletter.join(defletter).collect()
```
